# Data Cleaning

```python
# Drop NAs
clean_df = df.dropna()
clean_df.tail()
```

# Data Analysis

Find the graduate with the highest starting median salary

```python
idmax = clean_df['Starting Median Salary'].idxmax()
clean_df.loc[idmax]
```

```python
clean_df['Undergraduate Major'].loc[idmax]
```

Find the graduate with the lowest mid-career median salary

```python
clean_df.iloc[clean_df['Mid-Career Median Salary'].idxmin()]
```

Find the graduate with the lowest starting median salary

```python
clean_df.iloc[clean_df['Starting Median Salary'].idxmin()]
```

Find the mid-career salary spread

```python
spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1, 'Spread', spread_col)
clean_df.head()
```

Majors with the lowest associated financial risk 

```python
low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

Majors with the highest associated financial risk

```python
high_spread = clean_df.sort_values('Spread', ascending=False)
high_spread[['Undergraduate Major', 'Spread']].head()
```

Majors with the highest potential

```python
highest_potential = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
highest_potential[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

Salary information by major groups (Business vs. HASS vs. STEM)

```python
grouped_df = clean_df.groupby('Group')
```

```python
pd.options.display.float_format = '{:,.2f}'.format 
grouped_df.mean()
```
